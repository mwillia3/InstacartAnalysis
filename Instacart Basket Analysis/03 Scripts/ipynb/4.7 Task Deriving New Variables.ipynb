### 1. Create 'Price Label' Column

```python
#Define a function using user-defined functions

def price_label(row):
    
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High Range'
    else: return 'Not enough data'
```

```python
#Check Output
df.shape
```

```python
df.head()
```

```python
#Apply the function

df['price_range'] = df.apply(price_label, axis = 1)
```

```python
#Check Output
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

```python
#Define function using loc() function
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

### 2. Create 'Busiest Day' Column

```python
#Check Outputs
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
#Assign Busiest Days and Least Busiest Days
result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
    if value == 0:
        result.append("Busiest Day")
    elif value == 4:
        result.append("Least busy")
    else:
        result.append("Regularly busy")
```

### 3. (Step 2) Create 'Busiest Days' Column

```python
#Assign Busiest Days and Least Busiest Days
result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
    if value == 0 or value == 1:
        result.append("Busiest Days")
    elif value == 4 or value == 3:
        result.append("Least busy")
    else:
        result.append("Regularly busy")
```

```python
result
```

```python
#Create "Busiest Days" Column
df_ords_prods_merge['busiest_days'] = result
```

```python
#Check Outputs
df_ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

Busiest days should be the sum of days "0" and "1". It is.
Least busy days should be the sum of days "4" and "3". It is. 
Regularly busy days should be the sum of days "2", "5", and "6". It is. The loop was operated correctly.

### 4. (Step 4 & 5) Create 'Busiest Period of Day' Column

```python
#Check Outputs
df_ords_prods_merge.head()
```

```python
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

There are 24 hours in the day. If the hours are split equally into the categories "Most Orders", "Average Orders", and "Fewest Orders", the hours would line up as:
"Most Orders" = 10, 11, 14, 15, 13, 12, 16, 9
"Average Orders" = 17, 8, 18, 19, 20, 7, 21, 22
"Fewest Orders" = 23, 6, 0, 1, 5, 2, 4, 3

```python
#Assign Busiest Hours and Least Busiest Houses
result = []

for value in df_ords_prods_merge["order_hour_of_day"]:
    if value in [10, 11, 14, 15, 13, 12, 16, 9]:
        result.append("Most Orders")
    elif value in [17, 8, 18, 19, 20, 7, 21, 22]:
        result.append("Average Orders")
    elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
        result.append("Fewest Orders")
    else:
        result.append("Not Enough Data")
```

```python
result
```

```python
#Create "Busiest Period of Day" Column
df_ords_prods_merge['busiest_period_of_day'] = result
```

```python
#Check Outputs
df_ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```

The grouped sums match the individual times of day added together showing that the loop was operated correctly. Also, there are no rows that have missing data also showing it was done correctly.
